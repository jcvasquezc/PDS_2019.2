#### Agregar elementos a una lista

Elementos pueden er agregados a una lista usando el método ```append()```

Solo un elemento a la vez puede ser agregado. Para la inclusión de multiples elementeos usado la función ```append()```, se deben usar ciclos ```for```

```append()``` solo funciona para adicionar un elemento al final de la lista. Para agregar elementos en una posicón específica, se usa el método ```insert()```. 

Existe otro metodo llamado ```extend()```, el cual es utilizado para gregar multiples elementos al final de la lista.

````python
# Creating a List 
List = [] 
print("Lista vacia al inicio: ") 
print(List) 
  
# Addition of Elements  
# in the List 
List.append(1) 
List.append(2) 
List.append(4) 
print("\nLista despues de agregar 3 elementos: ") 
print(List) 
  
# Adding elements to the List 
# using Iterator 
for i in range(1, 4): 
    List.append(i) 
print("\nOtra forma de agregar elementos a una lista, numeros del 1 al 3: ") 
print(List) 
  
# Adding Tuples to the List 
List.append((5, 6)) 
print("\nSe adiciona una tupla a la lista: ") 
print(List) 
  
# Addition of List to a List 
List2 = ['de ', 'Senales'] 
List.append(List2) 
print("\nLista despues de adicionar una lista: ") 
print(List) 
  
# Addition of Element at  
# specific Position 
# (using Insert Method) 
List.insert(3, 12) 
List2.insert(0, 'Procesamiento') 
print("\nLista despues de una operación de Inserción: ") 
print(List) 
  
# Addition of multiple elements 
# to the List at the end 
# (using Extend Method) 
List.extend([8, 'Siempre', 'Señales']) 
print("\nLista despues de una operación de Extensión: ") 
print(List) 
````

### Acceder a elementos de una lista

Con el fin de acceder a u elemento de unna lista, se debe usar su **Indice**

Se usa el operador de indice [ ] para acceder a un item en una lista.
El indice debe ser un valor entero




````python
# Creating a List with 
# the use of multiple values 
List = ['Procesamiento', "de", "Señales"] 
  
# accessing a element from the  
# list using index number 
print("Acceder a los elementos de una lista") 
print(List[0])  
print(List[2])
  

List = [1, 2, 'procesamiento', 4, 'De', 6, 'Señales'] 
  
# accessing a element using 
# negative indexing 
print("Acceder a los elementos con indice negativo") 
  
# print the last element of list 
print(List[-1]) 
  
# print the third last element of list  
print(List[-3]) 
````

### Deslizamiento de listas

El deslizamiento se utiliza para acceder a elementos de una lista en un rango específico.

El deslizamiento se realiza con el operador ```:```

Para acceder a elementos desde el principio de la lista a un rango específico, se utiliza [:Index]

Para acceder a elementos desde el final de la lista a un rango específico, se utiliza [:-Index]

Para acceder a elementos desde  rango específico al final, se utiliza [Index:]

Para acceder a elementos dentro de un rango, se utiliza [Start Index:End Index] 

Además, para acceder a una Lista en orden inverso, se usa [::-1].

````python
# Creating a List 
List = ['G','E','E','K','S','F', 
        'O','R','G','E','E','K','S'] 
print("Lista Inicial: ") 
print(List) 
  
# Print elements of a range 
# using Slice operation 
Sliced_List = List[3:8] 
print("\nDeslizamiento de los indices 3-8: ") 
print(Sliced_List) 
  
# Print elements from beginning 
# to a pre-defined point using Slice 
Sliced_List = List[:-6] 
print("\nelementos hasta el sexto elemento desde el ultimo: ") 
print(Sliced_List) 
  
# Print elements from a  
# pre-defined point to end 
Sliced_List = List[5:] 
print("\nElementos desde el quinto"
      "hasta el final: ") 
print(Sliced_List) 
  
# Printing elements from 
# beginning till end 
Sliced_List = List[:] 
print("\nTodos los elementos de la lista: ") 
print(Sliced_List) 
  
# Printing elements in reverse 
# using Slice operation 
Sliced_List = List[::-1] 
print("\nElementos de la lista en orden inverso: ") 
print(Sliced_List) 
````

## Matplotlib

Grafica funciones sinusoidales difindas en un arreglo

````python
#t=np.linspace(0,20,10)
t=np.linspace(0,20,1000)
print(t)

y1 = np.sin(t)
y2 = np.cos(t)

print(y1)
print(y2)
````

````python
import matplotlib.pyplot as plt
````

````python
plt.plot(t , y1, "-g", label="sine")
plt.plot(t , y2, "-b", label="cos")

plt.legend(loc="upper right")

plt.ylim(-1.5, 1.5)
plt.show()
````

````python
f=1. # frecuencia de la senal sinusoidal
fs=30. # frecuencia de muestreo, cuantas muestras hay en un segundo
t=np.arange(0,4., 1./fs)
print(t)

x=np.sin(2*np.pi*f*t)
plt.plot(t,x, '-')
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
````
